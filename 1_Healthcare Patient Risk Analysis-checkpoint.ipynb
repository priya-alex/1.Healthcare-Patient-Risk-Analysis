```python
# ===================================================
# FULL SYNTHETIC PATIENT RISK PIPELINE (Single Jupyter Cell)
# - Outputs saved in a single folder: 'patient_outputs'
# ===================================================

import os, sys, subprocess
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.cluster import KMeans
import joblib

# --------------------------
# Helper: attempt pip install (silent)
# --------------------------
def ensure_package(pkg_name):
    try:
        __import__(pkg_name)
        return True
    except Exception:
        try:
            print(f"Installing {pkg_name} ...")
            subprocess.check_call([sys.executable, "-m", "pip", "install", pkg_name], stdout=subprocess.DEVNULL)
            __import__(pkg_name)
            return True
        except Exception as e:
            print(f"Could not install {pkg_name}: {e}")
            return False

# --------------------------
# Create output folder
# --------------------------
output_folder = "patient_outputs"
os.makedirs(output_folder, exist_ok=True)

# --------------------------
# 1) Generate synthetic dataset
# --------------------------
np.random.seed(42)
n_patients = 1000

df = pd.DataFrame({
    'patient_id': np.arange(1, n_patients + 1),
    'age': np.random.randint(20, 90, n_patients),
    'gender': np.random.choice(['Male', 'Female'], n_patients),
    'bmi': np.round(np.random.normal(25, 4, n_patients), 1),
    'blood_pressure_systolic': np.random.randint(90, 180, n_patients),
    'cholesterol': np.random.randint(150, 300, n_patients),
    'glucose': np.random.randint(70, 200, n_patients),
    'smoker': np.random.choice([0, 1], n_patients, p=[0.8,0.2]),
    'alcohol_use': np.random.choice([0, 1], n_patients, p=[0.85,0.15]),
    'exercise_level': np.random.choice(['Low', 'Medium', 'High'], n_patients, p=[0.3,0.5,0.2]),
    'past_admissions_12mo': np.random.poisson(0.8, n_patients),
    'readmitted_30d': np.random.choice([0,1], n_patients, p=[0.88,0.12]),
})

df['comorbidity_count'] = ((df['age']>65).astype(int) + 
                           (df['bmi']>30).astype(int) + 
                           (df['glucose']>125).astype(int) + 
                           (df['cholesterol']>=240).astype(int))

risk_score = (df['comorbidity_count'] * 0.6) + (df['past_admissions_12mo'] * 0.4) + ((df['age']>75).astype(int) * 1.2)
df['risk_level'] = pd.cut(risk_score, bins=[-1,0.5,1.5,10], labels=['Low','Medium','High'])

print("Synthetic dataset created:", df.shape)
display(df.head())

# --------------------------
# 2) Feature engineering
# --------------------------
# Age group
bins = [0, 35, 50, 65, 80, 200]
labels = ['18-34', '35-49', '50-64', '65-79', '80+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)

# BMI category
def bmi_cat(x):
    try: x = float(x)
    except: return 'Unknown'
    if x < 18.5: return 'Underweight'
    if x < 25: return 'Normal'
    if x < 30: return 'Overweight'
    return 'Obese'
df['bmi_category'] = df['bmi'].apply(bmi_cat)

# Lab flags
df['glucose_high_flag'] = (df['glucose'] > 125).astype(int)
df['chol_high_flag'] = (df['cholesterol'] >= 240).astype(int)

# Med adherence (synthetic)
df['med_adherence'] = np.random.choice([0.9, 0.8, 0.6, 0.95], n_patients)

# Comorbidity count
df['comorbidity_count'] = df['comorbidity_count'].fillna(0).astype(int)

# Simulate ALOS
df['length_of_stay'] = np.round(np.random.exponential(3, n_patients) + 2).astype(int)

# --------------------------
# 3) Prepare features & target
# --------------------------
target_col = 'risk_level'
num_cols = ['age','bmi','blood_pressure_systolic','cholesterol','glucose',
            'past_admissions_12mo','comorbidity_count','length_of_stay','med_adherence']
num_cols = [c for c in num_cols if c in df.columns]

cat_cols = [c for c in ['gender','exercise_level','bmi_category','age_group'] if c in df.columns]

X_num = df[num_cols].copy()
X_cat = pd.get_dummies(df[cat_cols].astype(str), drop_first=True) if cat_cols else pd.DataFrame(index=df.index)
X_all = pd.concat([X_num, X_cat], axis=1).fillna(0)

le_target = LabelEncoder()
y = le_target.fit_transform(df[target_col].astype(str))
print("Target classes:", list(le_target.classes_))
print("Feature matrix shape:", X_all.shape)

# --------------------------
# 4) Train/test split & scale
# --------------------------
X_train, X_test, y_train, y_test, idx_train, idx_test = train_test_split(
    X_all, y, df.index, test_size=0.20, stratify=y, random_state=42
)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# --------------------------
# 5) Train RandomForest
# --------------------------
rf = RandomForestClassifier(n_estimators=300, class_weight='balanced', random_state=42)
rf.fit(X_train_scaled, y_train)
y_pred = rf.predict(X_test_scaled)
y_proba = rf.predict_proba(X_test_scaled)

print("\nMODEL EVALUATION (RandomForest):")
print(classification_report(y_test, y_pred, target_names=list(le_target.classes_)))
print("Accuracy:", accuracy_score(y_test, y_pred))

# --------------------------
# 6) SHAP explainability
# --------------------------
shap_available = False
if ensure_package('shap'):
    try:
        import shap
        shap_available = True
    except:
        shap_available = False

if shap_available:
    try:
        explainer = shap.TreeExplainer(rf)
        X_shap = X_test_scaled[:300] if X_test_scaled.shape[0] > 300 else X_test_scaled
        shap_values = explainer.shap_values(X_shap)
        shap.summary_plot(shap_values, features=X_shap, feature_names=X_all.columns.tolist(), show=False)
        plt.tight_layout()
        plt.savefig(os.path.join(output_folder, "shap_summary.png"), dpi=150)
        plt.show()
        print("Saved SHAP summary plot")
    except:
        print("SHAP plotting failed  falling back to feature importances")
else:
    print("SHAP not available  using RF feature importances instead.")

# --------------------------
# 7) Feature importance
# --------------------------
feat_imp = pd.Series(rf.feature_importances_, index=X_all.columns).sort_values(ascending=False)
top_global = feat_imp.head(20)
plt.figure(figsize=(10,6))
sns.barplot(x=top_global.values, y=top_global.index)
plt.title("Top 20 Global Feature Importances (RandomForest)")
plt.tight_layout()
plt.savefig(os.path.join(output_folder, "top_feature_importances.png"))
plt.show()

top_global.reset_index().rename(columns={'index':'feature',0:'importance'}).to_csv(
    os.path.join(output_folder,"top_risk_factors.csv"), index=False
)

# --------------------------
# 8) Patient-level predictions
# --------------------------
X_all_scaled = scaler.transform(X_all)
all_pred = rf.predict(X_all_scaled)
all_proba = rf.predict_proba(X_all_scaled)

pred_labels = le_target.inverse_transform(all_pred)
scores_df = pd.DataFrame({'patient_id': df['patient_id'],'predicted_label': pred_labels})
for i, cls in enumerate(le_target.classes_):
    scores_df[f'prob_{cls}'] = all_proba[:, i]

top5 = feat_imp.index[:5].tolist()
for f in top5:
    if f in df.columns:
        scores_df[f"feat_{f}"] = df[f].values
    elif f in X_all.columns:
        scores_df[f"feat_{f}"] = X_all[f].values
    else:
        scores_df[f"feat_{f}"] = np.nan

# --------------------------
# 9) KMeans segmentation
# --------------------------
kmeans_k = 4
kmeans = KMeans(n_clusters=kmeans_k, random_state=42)
clusters = kmeans.fit_predict(X_all_scaled)
df['cluster'] = clusters
cluster_df = df[['patient_id','cluster']].copy()

# --------------------------
# 10) Save all outputs in single folder
# --------------------------
df.to_csv(os.path.join(output_folder, "cleaned_patient_data.csv"), index=False)
scores_df.to_csv(os.path.join(output_folder, "patient_risk_scores.csv"), index=False)
cluster_df.to_csv(os.path.join(output_folder, "patient_segments.csv"), index=False)
joblib.dump({'model': rf, 'scaler': scaler, 'label_encoder': le_target, 'features': X_all.columns.tolist()}, 
            os.path.join(output_folder, "best_model.joblib"))

print(f"\nAll outputs saved in folder: {output_folder}")

# --------------------------
# 11) Quick KPIs & plots
# --------------------------
print("\n--- Quick KPIs ---")
if 'readmitted_30d' in df.columns:
    rr = df['readmitted_30d'].mean() * 100
    print(f"Readmission Rate (%): {rr:.2f}")

print("Predicted Risk Distribution (%):")
print(scores_df['predicted_label'].value_counts(normalize=True).mul(100).round(2).to_string())

if 'length_of_stay' in df.columns:
    print(f"Average Length of Stay (ALOS): {df['length_of_stay'].mean():.2f}")

plt.figure(figsize=(6,4))
sns.countplot(x='predicted_label', data=scores_df, order=le_target.classes_)
plt.title("Predicted Risk Counts")
plt.tight_layout()
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='cluster', data=df)
plt.title("Cluster Counts")
plt.tight_layout()
plt.show()

print("\nOutputs ready in the folder:", output_folder)
```
